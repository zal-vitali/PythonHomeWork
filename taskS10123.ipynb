```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data20152019.csv', sep = ';', encoding='cp1251')
#df
```

Задача 1. Найдите, какая область центрального федерального округа имела наибольшую численность студентов вечерней формы обучения в 2015 году.

```python
df_cd = df.iloc[2:19,:] #строки без итогов по ЦФО и без г. Москва
max_val = df_cd['Численность студентов очно-заочная (вечерняя) форма, человек, 2015'].max()

max_df_cd = df_cd[df_cd['Численность студентов очно-заочная (вечерняя) форма, человек, 2015'] == max_val]
max_df_cd
```

Задача 2. Постройте диаграмму с данными о численности студентов дневной формы обучения южного федерального округа за 2017 год.

```python

#df_distr = df[df['Субъект РФ'].str.contains(r'\bокруг')] #(почему-то не работает отбор)
#df_distr

df_south_distr = df.iloc[33:41,:]
plot = sns.barplot(df_south_distr, x = 'Субъект РФ', y = 'Численность студентов, очная форма, человек, 2017')
plot.set_xticklabels(labels = df_south_distr['Субъект РФ'], rotation = 90)
```

Задача 3. Постройте диаграмму количества студентов заочной формы обучения за 2019 год по всем регионам, в которых общее количество студентов не превышает 10000 за данный год.

```python
df_total_2019 = df[df['Численность студентов, всего человек, 2019']<=10000]
msk = df_total_2019['Субъект РФ'].str.contains('округ')
df_total_2019 = df_total_2019.drop(df_total_2019[msk].index)
plot = sns.barplot(df_total_2019, x = 'Субъект РФ', y = 'Численность студентов заочная форма, человек, 2019')
plot.set_xticklabels(labels = df_total_2019['Субъект РФ'], rotation = 90)
```
