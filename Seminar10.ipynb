```python
# df.drop(2) #Удалили строку 2
df = pd.DataFrame(df, columns=['Субъект РФ',
                               'Численность студентов, очная форма, человек',
                               'Год формирования сведений'])
#df
```

```python
msk = df['Субъект РФ'].str.contains('округ')
#msk
df = df.drop(df[msk].index)
#df
```

Составьте столбчатую диаграмму для
данных, полученных в задаче 2.
Среди имеющихся субъектов, найдите субъекты с
максимальным и минимальным количеством
студентов на очной-форме обучения.

```python
min_val = df['Численность студентов, очная форма, человек'].min()
min_val

min_df = df[df['Численность студентов, очная форма, человек'] == min_val]
min_df
```

```python
max_val = df['Численность студентов, очная форма, человек'].max()
max_val

max_df = df[df['Численность студентов, очная форма, человек'] == max_val]
max_df
```
